Pada kali ini saya akan mencoba menganalisa salah satu series tv yaitu F.R.I.E.N.D.S, untuk menganalisa saya mendapatkan data dari kaggle, tujuan dari analisa ini untuk mencari tahu beberapa informasi atau fakta yang menarik yang terdapat dalam series F.R.I.E.N.D.S

```python
import pandas as pd
import numpy as np 

#visual
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Get Data


```python
data = pd.read_csv('imdb', parse_dates=['original_air_date'])
episode = pd.read_csv('episodes.csv', parse_dates=['original_air_date'])
```

```python
episode.head()
```

```python
data.head()
```

# Data wrangling

dari table diatas terdapat 2 file dengan informasi yang berbeda, untuk ini saya akan menggambungkan kedua file tersebut menjadi satu.

```python
data = data[['title', 'original_air_date', 'imdb_rating', 'total_votes', 'desc']]

all_df = episode.merge(data, how='left', on=['title', 'original_air_date'])

all_df
```

# EDA

Dari Data diatas kita mendapatkan beberapa informasi, yang dapat dijadikan beberapa jawab atas pertanyaan:
1. US Views terbanyak dan tersedikit
2. Top 5 episode berdasarkan IMDB Rating
3. Director Favorit berdasarkan jumlah episode dan IMDB Rating

# 1.US Views terbanyak dan tersedikit

```python
avg_view=all_df.us_viewers.mean()
max_view=all_df.us_viewers.max()
min_view=all_df.us_viewers.min()
print('Rata-rata views yang di dapat dari seluruh video :','%d'%(avg_view),'\nviews tersedikit dari satu video yaitu :',min_view,'\nviews terbanyak dari satu video yaitu :',max_view)
```

```python
views = all_df[['title','us_viewers']].sort_values('us_viewers', ascending=False).tail(10).reset_index(drop=True)
views
```

```python
top_views = all_df[['title','us_viewers']].sort_values('us_viewers', ascending=False).head(10).reset_index(drop=True)
top_views
```

```python
rat_eps5 = all_df[['title','imdb_rating']].sort_values('imdb_rating', ascending=False).head(5).reset_index(drop=True)
rat_eps5
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y=rat_eps5.title, x=rat_eps5.imdb_rating, palette='Blues_d')
plt.title('Top 5 High-rated Episodes', fontsize=15)
plt.xlabel('IMDB Stars', fontsize=13)
plt.ylabel('Episode', fontsize=13)
plt.xticks(np.arange(9, 9.8, 0.1), fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(9, 9.8)
```

```python
directed = all_df.directed_by.value_counts().to_frame().reset_index()
directed.columns = ['directed_by', 'title']
direct_rating = all_df.groupby('directed_by').imdb_rating.sum().to_frame().sort_values('imdb_rating', ascending=False).reset_index()
```

```python
direct = direct_rating.merge(directed, how='left', on='directed_by')
direct['Average'] = round(direct.imdb_rating / direct['title'], 2)
direct_plt = direct[direct['title']>=10].sort_values('Average', ascending=False)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y=direct_plt.directed_by, x=direct_plt.Average, palette='gist_gray_r')
plt.title('Top 8 High-rated Directors (Directing more than 10 Episodes)', fontsize=15)
plt.xlabel('Average IMDB Rating of the Episodes', fontsize=13)
plt.ylabel('Director', fontsize=13)
plt.xticks(np.arange(8, 8.8, 0.1), fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(8, 8.7)
```
