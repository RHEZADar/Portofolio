1. Mencari penjualan yang dilakukan pada setelah tanggal 1997-07-04

```python
sql(
"""
SELECT * 
FROM orders 
WHERE order_date > '1997-07-04'
""")
```

2. Jumlah Pembeli Berdasarkan Negara asal

```python
sql("""
select ship_country, count(order_id)
from orders
group by ship_country
order by count(order_id) desc
""")
```

3. Berapa Jumlah unit stock yang ada dan berikan keterangan untuk unit 0 diberikan keterangan 'Restock" dan lebih dari 0 diberikan "Ready"

```python
sql(
"""
SELECT product_name, Sum(units_in_stock) as units_in_stock,
CASE
WHEN sum(units_in_stock) = 0 THEN 'Restock'
ELSE 'Ready'
END as remarks
FROM products
group by product_name
order by units_in_stock
"""
)
```

4. Membuat data frame dengan menggunakan inner join

```python
data = sql(
"""
SELECT orders.order_id,orders.order_date, customers.company_name,order_details.unit_price, order_details.quantity, order_details.discount, products.product_name,
orders.ship_country
from orders
inner join order_details
on orders.order_id = order_details.order_id
inner join products
on order_details.product_id = products.product_id
inner join customers 
on orders.customer_id = customers.customer_id
"""
)
```
